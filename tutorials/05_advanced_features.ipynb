## 2. WebSocket Streaming

Get real-time updates during debates via WebSocket.

```python
# WebSocket event types
events = [
    ("debate_start", "Debate has begun"),
    ("round_start", "New round starting"),
    ("agent_message", "Agent is responding"),
    ("critique", "Agent is critiquing"),
    ("vote", "Agent has voted"),
    ("consensus", "Consensus reached"),
    ("debate_end", "Debate complete")
]

print("WebSocket Events:")
print("=" * 60)
for event, desc in events:
    print(f"  {event}: {desc}")
```

```python
# JavaScript client example
js_code = """
// Connect to WebSocket
const ws = new WebSocket('ws://localhost:8765/ws');

ws.onmessage = (event) => {
    const data = JSON.parse(event.data);
    console.log(data.type, data.payload);
    
    switch(data.type) {
        case 'agent_message':
            updateAgentPanel(data.payload);
            break;
        case 'consensus':
            showConsensus(data.payload);
            break;
    }
};

// Subscribe to a debate
ws.send(JSON.stringify({
    type: 'subscribe',
    debate_id: 'debate-123'
}));
"""

print("WebSocket Client Example:")
print(js_code)
```

## 3. ELO Tournaments

Run tournaments to rank agents by skill.

```python
from aragora.ranking.elo import ELOTracker, run_tournament

# Create ELO tracker
elo = ELOTracker(db_path="tutorial_elo.db")

# Run a tournament
print("Running ELO tournament...")
print("This pits agents against each other on standard tasks.")

# Tournament configuration
tournament_config = {
    "agents": ["anthropic-api", "openai-api", "gemini", "grok"],
    "tasks": [
        "Design a rate limiter",
        "Implement a cache strategy",
        "Review this security policy"
    ],
    "rounds_per_match": 2,
    "matches_per_pairing": 3
}

print(f"\nTournament: {len(tournament_config['agents'])} agents, {len(tournament_config['tasks'])} tasks")
```

```python
# Get current rankings
rankings = elo.get_rankings()

print("Current ELO Rankings:")
print("=" * 40)
print(f"{'Agent':25} {'ELO':>10}")
print("-" * 40)
for agent, rating in rankings[:10]:
    print(f"{agent:25} {rating:>10.0f}")
```

```python
# Get agent performance over time
history = elo.get_agent_history("anthropic-api")

print("\nAnthropic API History:")
print(f"  Matches played: {history.get('matches', 0)}")
print(f"  Win rate: {history.get('win_rate', 0):.0%}")
print(f"  Current rating: {history.get('rating', 1200):.0f}")
```

## 6. Batch Processing

Process multiple debates in parallel.

```python
# Batch input format (JSONL)
batch_example = """
{"question": "Design a rate limiter", "agents": "anthropic-api,openai-api"}
{"question": "Implement caching", "agents": "anthropic-api,gemini", "rounds": 4}
{"question": "Security review", "priority": 10}
"""

print("Batch Processing:")
print("=" * 60)
print("\n# Create a batch file (debates.jsonl):")
print(batch_example)

print("\n# Process locally:")
print("aragora batch debates.jsonl --output results.json")

print("\n# Submit to server:")
print("aragora batch debates.jsonl --server --wait")
```

## 7. Custom Convergence Detection

Configure how Aragora detects when agents are converging on a consensus.

```python
from aragora import DebateProtocol

# Advanced protocol configuration
protocol = DebateProtocol(
    rounds=5,
    consensus="majority",
    early_stopping=True,
    
    # Convergence detection
    convergence_detection=True,
    convergence_threshold=0.85,  # Stop when 85% similar
    
    # Vote grouping (cluster similar votes)
    vote_grouping=True,
    
    # Role rotation (agents switch roles each round)
    role_rotation=True,
    
    # Trickster detection (detect hollow consensus)
    enable_trickster=True
)

print("Advanced Protocol Configuration:")
print(f"  Convergence threshold: {protocol.convergence_threshold}")
print(f"  Role rotation: {protocol.role_rotation}")
print(f"  Trickster detection: {protocol.enable_trickster}")
```
